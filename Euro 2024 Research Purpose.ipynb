```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
euro = pd.read_csv("euro2024_players.csv")
```

```python
euro.head()
```

```python
average_age = np.mean(euro.Age)
print("Average Age: {} years".format(round(average_age,2)))
```

```python
average_height = np.mean(euro.Height)
print("Average height: {}cm".format(round(average_height,2)))
```

```python
clubs = euro.groupby("Club", as_index = False).agg({"Name":"count"})
clubs.columns = ["Club", "Count"]
clubs_sorted = clubs.sort_values(by = "Count", ascending = False).reset_index()
print(clubs_sorted)
```

```python
plt.figure(figsize = (8,7))
ax = plt.subplot()
plt.barh(clubs_sorted[clubs_sorted["Count"] >= 5].Club, clubs_sorted[clubs_sorted["Count"] >= 5].Count)
plt.tick_params("y", labelsize = 9)
ax.invert_yaxis()
plt.show()
```
